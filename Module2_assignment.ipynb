```python
# Q1 - String Operation
name = input("Enter your name: ")
print(f"Your name is:{name}")

print(f"Using title:{name.title()}")
print(f"First ch:{name[0].upper()} & Last ch: {name[-1].upper()}")
print(f"Reversed:{name[::-1]}")
print(f"Counting 'a':{name.count('a')}")
print(f"After removing space:{name.strip().replace(" ", "")}")
```

Reversed: Start and end calculations are as normal. From zero to len()-1. But -1 step indicates it to iterate backward. last index to first index.

```python
#Q5
n = int(input("Enter any number: "))
if n > 0:
  if n % 2 == 0:
    print(f"{n} is a Positive even number.")
  else:
    print(f"{n} is a Positive odd number.")

elif n < 0:
  if n % 2 == 0:
    print(f"{n} is a Negative even number.")
  else:
    print(f"{n} is a Negative odd number.")

else:
  print("Your enter value zero!")
```

```python
#Q6
x = float(input("Enter your physics mark: "))
y = float(input("Enter your mathematics mark: "))
z = float(input("Enter your chemistry mark: "))

total_marks = x + y + z
avg_mark = float(total_marks / 3)

print(f"Total: {total_marks}")
print(f"Average: {avg_mark}")

if x >= 40 and y >= 40 and z >= 40:
  print("Pass")
else:
  print("Fail.")

if avg_mark >= 75 and x >= 40 and y >= 40 and z >= 40:
  print("Distinction")
elif avg_mark >= 75 and (x < 40 or y < 40 or z < 40):
  print("Average is in Distinction level but result is fail")
elif avg_mark < 75 and (x < 40 or y < 40 or z < 40):
  print("Average is in below Distinction level also result is fail")
elif avg_mark < 75 and (x >= 40 or y >= 40 or z >= 40):
  print("Average is in below Distinction level but result is pass")
else:
  print("Normal Pass")
```

```python
nums = [12, 4, 8, 1, 9, 3, 4, 4]
nums.sort()
print(f"Sorted list: {nums}")

nums.insert(2, 100)
print(f"After inserting 100 at index 2: {nums}")

nums.remove(1)
print(f"After the removing the value 1: {nums}")

print(f"Maximum Value: {max(nums)}")
print(f"Minimum Value: {min(nums)}")

nums.reverse()
print(f"Reversed list: {nums}")

counts = nums.count(4)
print(f"Number of 4's: {counts} times.")
```
