```python
def process_scores(scores, threshold):
    lst1 = []
    lst2 = []

    for score in scores:
        if score > threshold:
            lst1.append(score)
        elif score <= threshold:
            lst2.append(score)

    return lst1, lst2


scores = [45, 60, 75, 30, 90, 50]
threshold = 50

above_list, below_or_equal_list = process_scores(scores, threshold)

print("Scores above threshold:", above_list)
print("Scores below or equal to threshold:", below_or_equal_list)
```

```python
def convert_temp(value, to_unit="C"):
    if to_unit == "C":
        return (value - 32) * 5 / 9
    elif to_unit == "F":
        return (value * 9 / 5) + 32
    else:
        raise ValueError("to_unit must be either 'C' or 'F'")

# Using positional arguments
f_value = 98.6
c_converted = convert_temp(f_value, "C")   # Fahrenheit to Celsius
print("Positional ->", f_value, "F =", round(c_converted, 2), "C")

# Using keyword arguments
c_value = 37
f_converted = convert_temp(value=c_value, to_unit="F")  # Celsius to Fahrenheit
print("Keyword   ->", c_value, "C =", round(f_converted, 2), "F")
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        return {
            "success": True,
            "result": result
        }

    except ZeroDivisionError:
        return {
            "success": False,
            "error": "Cannot divide by zero"
        }

    except TypeError:
        return {
            "success": False,
            "error": "Both a and b must be numbers"
        }

print(safe_divide(10, 2))       # valid
print(safe_divide(5, 0))        # division by zero
print(safe_divide("10", 2))     # type error
print(safe_divide(10, "2"))
```

```python
class BankAccount:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self.balance:
            print(f"Insufficient funds! Available balance: {self.balance}")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def get_balance(self):
        return self.balance


account = BankAccount("Alice", 1000)

print(f"Owner: {account.owner}, Starting balance: {account.get_balance()}")

account.deposit(500)
account.withdraw(200)
account.withdraw(2000)    # overdraft prevented
account.deposit(300)

print("Final balance:", account.get_balance())
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average(self):
        total = sum(self.marks.values())
        count = len(self.marks)
        if count == 0:
            return 0
        return total / count

student1 = Student("Alice", {"Math": 85, "English": 90, "Physics": 80})
student2 = Student("Bob",   {"Math": 70, "English": 75, "Biology": 65, "History": 80})

print(f"{student1.name}'s average score:", student1.average())
print(f"{student2.name}'s average score:", student2.average())
```

```python
import os
from datetime import datetime


class InventoryReport:
    def __init__(self, item_name, stock):
        self.item_name = item_name
        self.stock = stock


def generate_inventory_summary(report_obj, filename="inventory_report.txt", header="INVENTORY SUMMARY"):
    try:
        if not isinstance(report_obj, InventoryReport):
            raise TypeError("Object is not an InventoryReport instance")

        total_stock = sum(report_obj.stock.values())

        now = datetime.now()
        timestamp = now.strftime("%Y-%m-%d %H:%M:%S")

        with open(filename, "w") as f:
            f.write(header + "\n")
            f.write("=" * len(header) + "\n")
            f.write(f"Generated at: {timestamp}\n\n")
            f.write(f"Item: {report_obj.item_name}\n\n")
            f.write("Location-wise stock:\n")
            for location, count in report_obj.stock.items():
                f.write(f"  - {location}: {count}\n")
            f.write(f"\nTotal stock: {total_stock}\n")

        return os.path.abspath(filename)

    except Exception as e:
        return f"Error: {e}"

stock_data = {
    "Warehouse A": 120,
    "Warehouse B": 75,
    "Shop Floor": 30,
}

report = InventoryReport("Laptop", stock_data)

# 1) Call using positional arguments
path1 = generate_inventory_summary(report, "inv_positional.txt", "INVENTORY SUMMARY (POS)")
print("Positional call path:", path1)

# 2) Call using keyword arguments
path2 = generate_inventory_summary(
    report_obj=report,
    filename="inv_keyword.txt",
    header="INVENTORY SUMMARY (KEYWORD)"
)
print("Keyword call path:", path2)

# 3) Call with an invalid object to trigger error handling
invalid_result = generate_inventory_summary("not_a_report")
print("Invalid call result:", invalid_result)
```
