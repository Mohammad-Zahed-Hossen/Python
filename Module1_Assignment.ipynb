# Simple Calculator Program

```python
import math
x = int(input("Enter 1st input: "))
y = int(input("Ënter 2nd input: "))

# Mathematical Calculation (Addition, Subtraction, Multiplication, Division, Integer Division, Square Root)
print(f"Addition: {x + y}")
print(f"Subtraction: {x - y}")
print(f"Multiplication: {x * y}")
print(f"Division: {round(x / y, 2)}") #2 decimal point porjntw floating value k rounding korchi
print(f"Integer Division: {x // y}")
print(f"Remainder of {x} & {y}: {x % y}")
print(f"Square of {x}: {pow(x, 2)}")
print(f"Square Root of {x}: {round(math.sqrt(x), 2)}") #2 decimal point porjntw floating value k rounding korchi
```

# Temperature Converter

```python
temp = float(input("Enter temperature in Celsius: ")) #typecasting input temperature value into float
faren = (temp * float(9/5)) + 32

print(f"Temperature({temp}°C ) in Fahrenheit is: {faren}°F")
```

# Billing Program

```python
item_name = input("Enter item name: ")
price = float(input("Enter price of item:"))
quantity = int(input("Enter Quantity: "))

total = price * quantity
print(f"Total Bill for {item_name} = {total}")
```
